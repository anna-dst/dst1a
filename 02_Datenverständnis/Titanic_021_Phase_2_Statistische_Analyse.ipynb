Wir führen die Trainings- und Testdaten zusammen, sehen uns die Daten an und berechnen einige statistische Größen.

```python
# Pandas Paket (Package) importieren
#  Datenstrukturen und Datenanalyse, I/O
#  https://pandas.pydata.org/pandas-docs/stable/
import pandas as pd
```

```python
# Trainings- und Testdaten als Pandas Data Frame (df) aus CSV-Dateien laden
#  (KNIME: "CSV Reader")
df_train = pd.read_csv('../../data/titanic/original/train.csv')
df_test  = pd.read_csv('../../data/titanic/original/test.csv')
```

```python
# Daten zusammenführen
#  (KNIME "Concatenate")
# Hierzu benutzen wir die Funktion concat(),
# wobei wir eine Liste mit den einzelnen Data Frames als Argument übergeben.
df = pd.concat([df_train, df_test], ignore_index=True)
```

```python
# Statistische Informationen
#  (KNIME: "Statistics")
df.describe()
```

```python
# Fehlende Werte
#  (KNIME: "Statistics")
df.isnull().sum()
```

```python
# Datentypen bestmöglich automatisch konvertieren: z.B. object => string
df = df.convert_dtypes()
# Ausgabe
print(df.dtypes)
```

```python
# Überblick: Datentypen und Anzahl nicht-fehlender Daten anzeigen
df.info()
```

```python
# Die ersten und letzten Datensätze als formatierte Tabelle anzeigen
display(df)
```

### Ergebnisse der Datensichtung und der statistischen Analysen

Beschreibungen (siehe auch Kaggle: https://www.kaggle.com/c/titanic/data):

| Attribut    | Beschreibung                       | Domäne         | Skalenniveau | Datentyp | Fehlende Werte |
| ------------| -----------------------------------| ---------------| -------------| ---------| ---------------|
| PassengerId | Eindeutige Nummer                  | 1 bis 1309     | nominal      | Int      |                |
| Survived    | Überlebt                           | 0 oder 1       | nominal      | Int      | Testdaten: 418 |
| Pclass      | Passagierklasse                    | 1, 2, 3        | ordinal      | Int      |                |
| Name        | Passagiername                      | Text           | nominal      | String   |                |
| Sex         | Geschlecht                         | female, male   | nominal      | String   |                |
| Age         | Alter in Jahren                    | 0,17 bis 80    | kardinal     | Float    | 263            |
| SibSp       | # Geschwister & Ehepartner an Bord | 0 bis 8        | kardinal     | Int      |                |
| Parch       | # Eltern & Kinder an Bord          | 0 bis 9        | kardinal     | Int      |                |
| Ticket      | Ticketnummer                       | alphanumerisch | nominal      | String   |                |
| Fare        | Ticket-Preis in Pfund              | 0 bis 512,3292 | kardinal     | Float    | 1              |
| Cabin       | Kabinennummer                      | alphanumerisch | nominal      | String   | 1014           |
| Embarked    | Abfahrtshafen                      | C, Q, S        | nominal      | String   | 2              |

Legende mit Bedeutungen der Werte:
* Survived: 0 = Nein, 1 = Ja
* Pclass: 1 = erste, 2 = zweite, 3 = dritte 
* Embarked: C = Cherbourg, Q = Queenstown, S = Southampton

Erste Erkenntnisse:

* Die PassengerId ist ein Schlüsselattribut zur eindeutigen Identifizierung eines Passagiers. Es enthält aber keine Informationen, die für die Datenanalyse relevant sind.

* In den Namen stecken weitere Informationen wie z.B. die Anrede, der Titel, der Mädchenname (bei Frauen) usw.

* Das Alter wurde manchmal nur geschätzt (Kommazahl) und fehlt auch sehr häufig.

* Die genaue Bedeutung der Ticketnummer ist unklar.

* Der Ticket-Preis bezieht sich ggf. auf mehrere Passagiere, die unter der gleichen Ticketnummer gereist sind.

* Die Kabinennummer, falls vorhanden, beginnt mit dem Deck. In der 3. Klasse haben sich vermutlich viele Passagiere eine gemeinsame Kabine geteilt. Diese Daten sind kaum erfasst.
