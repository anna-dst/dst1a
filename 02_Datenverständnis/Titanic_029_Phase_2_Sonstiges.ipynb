```python
# Pandas Paket (Package) importieren
#  Datenstrukturen und Datenanalyse, I/O
#  https://pandas.pydata.org/pandas-docs/stable/
import pandas as pd
```

```python
# Trainings- und Testdaten als Pandas Data Frame (df) aus CSV-Dateien laden
#  (KNIME: "CSV Reader")
df_train = pd.read_csv('../../data/titanic/original/train.csv')
df_test  = pd.read_csv('../../data/titanic/original/test.csv')
```

```python
# Daten zusammenführen
#  (KNIME "Concatenate")
df = pd.concat([df_train, df_test], ignore_index=True)
```

```python
# (1) Passagiere mit geschätztem Alter
df['Age'] = df['Age'].astype(str) # Umwandlung in String
df[df['Age'].str.endswith('.5')]  # Age muss mit .5 enden 
```

```python
# (2) Passagiere mit langen (bzw. kurzen) Namen
nameLength = df['Name'].apply(len) # Länge des Namens berechnen
df.insert(loc=len(df.columns), column='NameLength', value=nameLength) # neue Spalte NameLength
df = df.sort_values(by=['NameLength'], ascending=False) # nach Namelength absteigend sortieren
pd.set_option('max_colwidth', 100) # Spaltenbreite für Ausgabe anpassen
display(df)
```
