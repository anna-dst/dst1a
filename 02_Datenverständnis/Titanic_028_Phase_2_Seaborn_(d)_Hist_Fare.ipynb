Wir erstellen ein Histogramm zum Attribut Fare -- mit Hilfe der Seaborn-Bibliothek.

```python
# Pandas Paket (Package) importieren
#  Datenstrukturen und Datenanalyse, I/O
#  https://pandas.pydata.org/pandas-docs/stable/
import pandas as pd
# Pyplot Modul der Matplot Bibliothek (Library) importieren
#  Mathematische grafische Daten-Visualisierungen
#  https://matplotlib.org/contents.html
import matplotlib.pyplot as plt
# Seaborn importieren [sns = seaborn name space, Samuel Norman Seaborn]
#  High-end Daten-Visualisierungen auf Basis von Matplot
#  https://seaborn.pydata.org/
import seaborn as sns
# Deutsches Zahlenformat
import locale
locale.setlocale(locale.LC_ALL, 'de')
# Warnungen unterdrücken
#import warnings
#warnings.filterwarnings('ignore')
```

```python
# Trainings- und Testdaten als Pandas Data Frame (df) aus CSV-Datei laden
#  (KNIME: "CSV Reader")
df_train = pd.read_csv('../../data/titanic/original/train.csv')
df_test  = pd.read_csv('../../data/titanic/original/test.csv')
```

```python
# Daten zusammenführen
#  (KNIME "Concatenate")
df = pd.concat([df_train, df_test], ignore_index=True)
```

```python
# Datentypen automatisch konvertieren
df_train = df_train.convert_dtypes()
df_test  = df_test.convert_dtypes()
df       = df.convert_dtypes()
```

```python
# Seaborn: Allgemeine Einstellungen

# Schriftart setzen
sns.set(font = 'Arial')
# Schriftgröße setzen
sns.set(font_scale = 2.5)
# Achsen im deutschen Zahlenformat
plt.rcParams['axes.formatter.use_locale'] = True
# Bildgröße setzen (16 x 9)
sns.set(rc={'figure.figsize':(8, 4.5)})
# Weißes Gitter setzen
sns.set_style('whitegrid')
```

```python
# Histplot zu Fare

# Fehlende Werte löschen
fare = df['Fare'].dropna().astype(float)

# stat = density, count, percent
sns.histplot(fare, kde=True, stat='count', kde_kws=dict(cut=3), alpha=.4, edgecolor=(1, 1, 1, .4), bins=40)

# Achsen beschriften
plt.xlabel('Reisepreis (Fare)')
plt.ylabel('Häufigkeit')

# Grafik speichern
plt.savefig('../../figures/svg/Hist_Fare.svg', format='svg', dpi=1200, transparent=True)
```
