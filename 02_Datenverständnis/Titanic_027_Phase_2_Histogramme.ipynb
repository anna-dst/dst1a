Histogramme sind spezielle Säulendiagramme, um Verteilungen zu visualisieren. Besonders interessant in dieser Fallstudie sind Histogramme bzw. Säulendiagramme, die die Überlebenswahrscheinlichkeiten zu den Titanic-Passagieren zeigen.

```python
# Pandas Paket (Package) importieren
#  Datenstrukturen und Datenanalyse, I/O
#  https://pandas.pydata.org/pandas-docs/stable/
import pandas as pd
# Pyplot Modul der Matplot Bibliothek (Library) importieren
#  Mathematische grafische Daten-Visualisierungen
#  https://matplotlib.org/contents.html
import matplotlib.pyplot as plt
```

```python
# Trainingsdaten als Pandas Data Frame (df) aus CSV-Datei laden
#  (KNIME: "CSV Reader")
df_train = pd.read_csv('../../data/titanic/original/train.csv')
```

```python
# Histogramm erstellen
#  (KNIME: "Histogram")
#
# Attribut Pclass
plt.hist(df_train['Pclass'])
# Überschrift und Achsenbeschriftungen
plt.title('Histogramm')
plt.xlabel('Passagierklasse (Pclass)')
plt.ylabel('Häufigkeit')
# Grafik ausgeben
plt.show()
```

```python
# Überlebenswahrscheinlichkeiten zum Attribut Pclass
probs = df_train.groupby(['Pclass'], as_index=False)['Survived'].mean()
display(probs)
```

```python
# Säulendiagramm erstellen
#  (KNIME: "Histogram")
#
plt.bar(probs['Pclass'], height=probs['Survived'], width=0.2, tick_label=probs['Pclass'])
# Überschrift und Achsenbeschriftungen
plt.title('Säulendiagramm')
plt.xlabel('Passagierklasse (Pclass)')
plt.ylabel('Überlebenswahrscheinlichkeit')
# Grafik ausgeben
plt.show()
```

```python
# Histogramm erstellen
#  (KNIME: "Histogram")
#
# Attribut Sex
plt.hist(df_train['Sex'])
# Überschrift und Achsenbeschriftungen
plt.title('Histogramm')
plt.xlabel('Geschlecht (Sex)')
plt.ylabel('Häufigkeit')
# Grafik ausgeben
plt.show()
```

```python
# Überlebenswahrscheinlichkeiten zum Attribut Pclass
probs = df_train.groupby(['Sex'], as_index=False)['Survived'].mean()
display(probs)
```

```python
# Säulendiagramm erstellen
#  (KNIME: "Histogram")
#
plt.bar(probs['Sex'], height=probs['Survived'], width=0.1, tick_label=probs['Sex'])
# Überschrift und Achsenbeschriftungen
plt.title('Säulendiagramm')
plt.xlabel('Geschlecht (Sex)')
plt.ylabel('Überlebenswahrscheinlichkeit')
# Grafik ausgeben
plt.show()
```

```python
# Histogramm erstellen
#  (KNIME: "Histogram")
#
# Attribut Age (10 Bins)
plt.hist(df_train['Age'], bins=10)
# Überschrift und Achsenbeschriftungen
plt.title('Histogramm')
plt.xlabel('Alter (Age)')
plt.ylabel('Häufigkeit')
# Grafik ausgeben
plt.show()
```

```python
# Überlebenswahrscheinlichkeiten zum Attribut Age (10 Bins)
df_train['AgeBinned'] = pd.cut(df_train['Age'].dropna(), bins=10)
probs = df_train.groupby(['AgeBinned'], as_index=False, observed=False)['Survived'].mean()
display(probs)
```

```python
# Säulendiagramm erstellen
#  (KNIME: "Histogram")
#
plt.bar(probs['AgeBinned'].astype(str), height=probs['Survived'])
# Überschrift und Achsenbeschriftungen
plt.title('Säulendiagramm')
plt.xlabel('Alter (Age)')
plt.ylabel('Überlebenswahrscheinlichkeit')
plt.xticks(rotation=45)
# Grafik ausgeben
plt.show()
```

```python
# Histogramm erstellen
#  (KNIME: "Histogram")
#
# Attribut Age (30 Bins)
plt.hist(df_train['Age'], bins=30)
# Überschrift und Achsenbeschriftungen
plt.title('Histogramm')
plt.xlabel('Alter (Age)')
plt.ylabel('Häufigkeit')
# Grafik ausgeben
plt.show()
```

Danach nimmt die Überlebenswahrscheinlichkeit teilweise auch wieder ab, allerding ebenso die statistische Grundlage.
=> Mit steigendem Ticket-Preis steigt zunächst auch die Überlebenschance.
