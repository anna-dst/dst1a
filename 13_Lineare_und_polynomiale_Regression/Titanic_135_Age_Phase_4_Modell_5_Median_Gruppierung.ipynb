Modell 5: Median des Alters (Age) gruppiert nach [Pclass, Sex, Title]

```python
# Pandas Paket (Package) importieren
#  Datenstrukturen und Datenanalyse, I/O
#  https://pandas.pydata.org/pandas-docs/stable/
import pandas as pd
# Klassen und Funktionen der scikit-learn Module importieren
#  Maschinelles Lernen, Algorithmen
#  https://scikit-learn.org/stable/modules/classes.html
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
# Trainingsdaten als Pandas Data Frame (df) aus Excel-Datei laden
#  (KNIME: "Excel Reader")
df_train = pd.read_excel('../../data/titanic/age/training_v1.xlsx')
```

```python
# Modell 5: Median nach Gruppierung 
#  (KNIME: "GroupBy")
median_age_group = df_train.groupby(['Pclass_2', 'Pclass_3', 'Sex_male', 'Title_Master', 'Title_Miss', 'Title_Mrs',
                          'Title_Rare'], as_index=False)['Age'].median()
median_age_group = median_age_group.rename(columns={'Age': 'Prediction'})
display(median_age_group)
```

```python
# Daten zusammenfügen
#  (KNIME: "Joiner")
df_train = df_train.merge(median_age_group, how='left', on=['Pclass_2', 'Pclass_3', 'Sex_male', 'Title_Master',
                                                            'Title_Miss', 'Title_Mrs', 'Title_Rare'])
display(df_train)
```

```python
# Gütemaße bestimmen und ausgeben
#  (KNIME: "Numeric Scorer")
#
r2   = r2_score(                      df_train['Age'], df_train['Prediction'])
rmse = root_mean_squared_error(       df_train['Age'], df_train['Prediction'])
mae  = mean_absolute_error(           df_train['Age'], df_train['Prediction'])
mape = mean_absolute_percentage_error(df_train['Age'], df_train['Prediction'])
#
print('R^2  : %8.4f' % (r2))
print('RMSE : %6.2f' % (rmse))
print('MAE  : %6.2f' % (mae))
print('MAPE : %5.1f %%' % (100*mape))
```
