```python
# Pandas Paket (Package) importieren
#  Datenstrukturen und Datenanalyse, I/O
#  https://pandas.pydata.org/pandas-docs/stable/
import pandas as pd
```

```python
# Anwendungsdaten als Pandas Data Frame (df) aus CSV-Datei laden
#  (KNIME: "CSV Reader")
filename = '../../data/titanic/original/application.csv'
df_app  = pd.read_csv(filename)
```

```python
# Modell 6: Frauen und Kinder der 1. und 2. Klasse überleben, alle anderen sterben. 
#  (KNIME: "Category To Number" und "Math Formula")
df_app['Prediction'] = ( ( (df_app['Sex'] == 'female') | (df_app['Age'] < 12) ) \
                       & ( (df_app['Pclass'] != 3) ) ).astype('int')
```

```python
# Ergebnisse filtern
#  (KNIME: "Column Filter")
df_res = df_app.filter(['PassengerId', 'Prediction'])
# Ergebnisse anzeigen
display(df_res)
```
